# Split Data

```python
features=df.drop('Result',axis=1)
target=df['Result']
```

```python
features.head()
```

```python
target.head()
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit(features)
```

```python
scaled_features=scaler.transform(features)
scaled_features
```

```python
type(scaled_features)
```

```python
#Before Scaling
df.head()
```

```python
col=df.columns[:-1]
col
```

```python
#After Scaling
df.scaled_features=pd.DataFrame(scaled_features,columns=col)
df.scaled_features
```

# Train Test split

```python
from sklearn.model_selection import train_test_split as tts
```

```python
x= features
y=target
```

```python
xtrain,xtest,ytrain,ytest=tts(x,y, test_size=.25, random_state=1)
```

```python
xtrain
```

```python
ytest
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=3)
```

```python
knn.fit(xtrain,ytrain)
```

```python
predictions=knn.predict(xtest)
predictions
```

```python
knn.score(xtest,ytest)
```
